```python
##Independent and dependent feature

X=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1:]
```

```python
#Train Test Split

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
##Standardising the dataset
#we standarise the data since there are multiple independent feature having differnt unit so we make it standard so that it's easier to apply gradient desent and reach global minima

from sklearn.preprocessing import StandardScaler
Scaler = StandardScaler ()
```

```python
X_train = Scaler.fit_transform(X_train)
```

```python
X_test = Scaler.fit_transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
##parameter on which model has been trained
regression.get_params()
```

```python
#predict with test data
reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

```python
##scatter plot
plt.scatter(y_test,reg_pred)
```

```python
##residual or error
residual = y_test - reg_pred
```

```python
##plot the residual
sns.displot(residual,kind="kde")
```

```python
##scatter plot wrt prediction, residual
#it's uniform districution
plt.scatter(reg_pred,residual)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test, reg_pred))
print(mean_squared_error(y_test, reg_pred))
print(np.sqrt(mean_squared_error(y_test, reg_pred)))
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,reg_pred)
score
```
